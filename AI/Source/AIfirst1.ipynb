```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df.head(3)
```

```python
df = pd.DataFrame.drop(df,columns = ["Person ID" , "Occupation", "Blood Pressure" , "Daily Steps" , "Sleep Disorder" , "Heart Rate" , "BMI Category"] , axis = "1")
```

```python
df = pd.get_dummies(df, columns = ["Gender"], drop_first=True)
```

```python
df.head(3)
```

```python
y = df[["Quality of Sleep" , "Stress Level"]]
x = df.drop(["Quality of Sleep" , "Stress Level"] , axis = "columns")
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
```

```python
model.predict([[17,7,40,True]])
```
